#### Обработка `year_of_release`

Год можно выявить из названия игры. Напишем функцию которая транслирует год из названия игры и применим её:

```python
data.loc[16711, 'name'].split()[-1]
def get_year_from_name(row):
    words = row['name'].split()
    if words[-1].isdigit() and 1980 <= int(words[-1]) < 2017:
        return int(words[-1])
    else:
        return np.nan
data.loc[data['year_of_release'].isna(), 'year_of_release'] = data.loc[data['year_of_release'].isna()].apply(get_year_from_name, axis=1)
display(f'Осталось пропусков в столбце "Год релиза": {data.year_of_release.isna().sum()}')
list(data.year_of_release)
```

Считаю не допустимым заполнение медианным значением оставшиеся пропуски, т.к. заполнение такого количества ячеек повлияет на конечные выводы.

```python
data.loc[~data.year_of_release.isna(), 'year_of_release'] = data['year_of_release'].astype('Int64')
```

```python
list(data.year_of_release)
```

Изменил на целочисленный тип, так как это категориальная переменная.

Количество пропусков было не велико. Скорее всего это человеческий фактор - забыли проставить год релиза.

#### Обработка `rating`
Некоторые отметки рэйтинга ESRB  устарели (`K-A`, `AO`) - изменим их названия на современные. Мы имеем мало значений для `EC` - присоединим их к категории `E`. Также, отметку `RP` заменим на np.nan - значит одно и тоже. Таким образом в столбце останутся четыре современные отметки и np.nan.

```python
data['rating'].unique()
```

```python
data.loc[data.rating == 'K-A', 'rating'] = 'EC'
data.loc[data.rating == 'AO', 'rating'] = 'M'
data.loc[data.rating == 'RP', 'rating'] = np.nan
data.loc[data.rating == 'EC', 'rating'] = 'E'
data.rating.unique()
```

Так как пропущенные значения мы никак не можем определить, объявим эти значения категорией `unknown`:

```python
data.loc[data.rating.isna(), 'rating'] = 'unknown'
```

Найдём сколько в среднем лет сущестовует игровая платформа. Под фразой "годы существования платформы" или "срок жизни платформы" будем понимать годы в которых имелись релизы игры для платформы. 

```python
platform_copy = list(platforms)
platform_copy.remove('PC')
```

```python
years = []
data_temp = data.loc[(data['name'] != 'Strongest Tokyo University Shogi DS')&(data['platform'] != 'PC')]
for platform in platform_copy:
    min_year = data_temp.loc[data['platform'] == platform, 'year_of_release'].min()
    max_year = data_temp.loc[data['platform'] == platform, 'year_of_release'].max()
    if (max_year == 2016 and (max_year - min_year) > 6) or max_year != 2016:
        years.append(max_year - min_year)
        display(f'{platform} exsisted {max_year - min_year}')
display(f'Среднее время существования платформы {round(sum(years)/len(years), 2)} лет' )
```

Рассмотрим через какой срок игровые гиганты начинают замещать поколения платформ на примере экосистем Xbox (XB, X360, XOne), PS (PS-PS4), Nintendo (NES, SNES, N64):

```python
pl_ps = ['PS', 'PS2', 'PS3', 'PS4']
years_ps = []
for pl in pl_ps:
    min_year = data.loc[data['platform'] == pl, 'year_of_release'].min()
    years_ps.append(min_year)
diff_ps=[]
for x, y in zip(years_ps[0::], years_ps[1::]):
    diff_ps.append(y-x)
display(diff_ps)
```

```python
pl_xb = ['XB', 'X360', 'XOne']
years_xb = []
for pl in pl_xb:
    min_year = data.loc[data['platform'] == pl, 'year_of_release'].min()
    years_xb.append(min_year)
diff_xb=[]
for x, y in zip(years_xb[0::], years_xb[1::]):
    diff_xb.append(y-x)
display(diff_xb)
```

```python
pl_n = ['NES', 'SNES', 'N64', 'GC', 'Wii', 'WiiU']
years_n = []
for pl in pl_n:
    min_year = data_temp.loc[data_temp['platform'] == pl, 'year_of_release'].min()
    years_n.append(min_year)
diff_n=[]
for x, y in zip(years_n[0::], years_n[1::]):
    diff_n.append(y-x)
display(diff_n)
```

```python
all_diff = diff_n + diff_xb + diff_ps
mean_diff = sum(all_diff) / len(all_diff)
display(f'Через {round(mean_diff, 2)} лет производители в среднем начинают замещать поколения платформ')
```

**Выводы**

1. В средней платформы существуют 9 c лишним лет, но уже через 6 лет начинается замещение платформы на платформу другого поколения. Как правило, дебют платформы нового поколения приходится на год сразу после года пиковых продаж старой платформы. 
2. Пик продаж приходится на вторую треть существования платформы. Т.е. продажи в течение существования платформы стремятся к нормальному распределению. Исключение есть: платформа `2600` имеет пик продаж в первой трети существования платформы.

**H1**: средние пользовательских рейтингов не равны или `xb_sample.mean()` - `pc_sample.mean()` != 0

Отвергнуть или принять нулевую гипотезу поможет t-тест. 

Рассмотрим, подходит ли он нам по его ограничениям:

1. Генеральные совокупности не должны зависеть друг от друга. Наши совокупности не зависимы.
2. Выборочные средние должны быть нормально распределены.Благодаря ЦПТ, если размер выборки составляет хотя бы несколько десятков значений, выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены нормально вокруг истинного среднего этой совокупности. Напомним, это утверждение верно, даже если сама генеральная совокупность не распределена нормально. По размеры выборок мы проходим.
3. Дисперсии рассматриваемых генеральных совокупностей должны быть равны. Так как мы имеем разные по размеры выборки, мы укажем в параметрах функции для расчёта equal_var равным False - это условия обойдём таким образом.

Для проверки гипотезы будем использовать критерий p-value - это вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна.

```python
results = st.ttest_ind(xb_sample, pc_sample, equal_var=False)
print(f'p-value: {results.pvalue}')
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**: 

 Мы имеем p-value: 0,29 - часть площади нормального распределения. 
 
 Вся площадь ограниченная линией нормального распределения сверху и осью абсцисс равна вероятности в 1, что средние наших выборок **не равны**, при условии, что нулевая гипотеза верна.
 
 Нормальное распределение имеет форму колокола с "хвостами" слева и справа.  
 Так как у нас двустороняя гипотеза, p-value показывает долю площади "хвостов", т.е каждый хвост имеет площадь 29.4%/2. "Хвосты" ограниченны разностью между `xb_sample.mean()` и `pc_sample.mean()` (она м.б. и отрицательной и положительной), точнее эквивалентами положительной и отрицательной разностями (получают с помощью z-критерия). 

  Критерий `alpha` показывает суммарную площадь хвостов нормального распределения выборочного среднего (суммируются площади  для обоих хвостов), которую мы приняли как статистически значимую.
  
  Мы не отвергаем нулевую гипотезу потому что наша `alpha` гораздо меньше 29.4%. Т.е. с вероятностью 29.4%, мы получим ошибку в утверждении "средние значения не равны" **при условии, что нулевая гипотеза верна**. 
  
  Замечание: что нулевая гипотеза не отвергается, не следует, что она принимается. Т.е. нулевая гипотеза не противоречит имеющимся данным.

Формулирование двусторонних гипотез. <br>
        
<b>Никакие экспериментально полученные данные никогда не подтвердят какую-либо гипотезу. Это наше фундаментальное ограничение. </b>Данные могут лишь не противоречить ей или, наоборот, показывать крайне маловероятные результаты (при условии, что гипотеза верна). Но и в том, и в другом случае нет оснований утверждать, что выдвинутая гипотеза доказана.
Допустим, данные гипотезе не противоречат, тогда мы её не отвергаем. Если же мы приходим к выводу, что получить такие данные в рамках этой гипотезы вряд ли возможно, у нас появляется основание отбросить эту гипотезу.
    
P-значение (англ. P-value) — величина, используемая при тестировании статистических гипотез. <b>Фактически это вероятность ошибки при отклонении нулевой гипотезы (ошибки первого рода)</b> пример ниже

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

Напомним, это утверждение верно, даже если сама генеральная совокупность не распределена нормально. По размеры выборок мы проходим.
3. Дисперсии рассматриваемых генеральных совокупностей должны быть равны. Так как мы имеем разные по размеры выборки, мы укажем в параметрах функции для расчёта equal_var равным False - это условия обойдём таким образом.

Для проверки гипотезы будем использовать критерий p-value - это вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна.

```python
results = st.ttest_ind(action_sample, sports_sample, equal_var=False)
print(f'p-value: {results.pvalue}')
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**: получили очень низкую вероятность ошибки в утверждении "две средних не равны". Это дает уверенность в том, что средние пользовательских рейтингов для совокупностей игр жанров `Action` и `Sports` не равны. Считаем результат статистически значимым, так как p-value сильно ниже установленного порога в 1%.
