# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

```python
# Напишем функцию, которая унифицирует название типов населённых пунктов
def get_unification_name(location):
    if 'городской поселок' in location:
        return location.replace('городской поселок', 'поселок городского типа')
    if 'городской посёлок' in location:
        return location.replace('городской посёлок', 'поселок городского типа')
    if 'посёлок' in location:
        return location.replace('посёлок', 'поселок')
    return location   
data['locality_name'] = data['locality_name'].apply(get_unification_name)
len(data['locality_name'].unique())
```

Изменили данные в столбце для действия с датами. 

Так как мы можем производить действия с датами, проверим предположение:

Если добавить дни из `days_exposition` к дате из `first_day_exposition` получим максимальную дату 3 мая 2019 года - это объяснит пропуски в столбце `days_exposition` и мы будем понимать, что на 3 мая 2019 года эти квартиры не были проданы.

```python
# Создадим столбец с датой продажи, прибавив к дате размещения объявления количество дней из 'days_exposition':
data['date_sale'] = data['first_day_exposition'] \
    + data['days_exposition'].apply(np.ceil).apply(lambda x: pd.Timedelta(x, unit='D'))
```

```python
# Для начала создадим функцию, которая возвращает расположения этажа
def get_floor_details(floor, floors_total):
    ''' Этажи поделим по категориям:
        1 — 'первый';
        2–предпоследний — 'другой';
        последний — 'последний';
    '''
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'

# Добавим новый столбец 'floor_details'
data['floor_details'] = data.apply(lambda x: get_floor_details(x['floor'], x['floors_total']), axis=1)
display(data['floor_details'].unique())
```

```python
# Создадим функцию для расчёта наибольшего (largest_observation), наименьшего наблюдения (smallest_observation):
def get_observation(data, course='largest'):
    iqr = data.quantile(0.75) - data.quantile(0.25)
    if course == 'largest':     
        return data.quantile(0.75) + 1.5 * iqr
    elif course == 'smallest':
        return data.quantile(0.75) - 1.5 * iqr
    
largest_total_area = get_observation(data['total_area'])
```

### Задание граничных параметров для нормализованной базы

Имея нормализованную базу, установим параметры для выявления рыночной стоимости жилья и выявление аномальных значений.

Создадим таблицу с ценой квартир с различным количеством комнат, в которой представлена рыночная цена в виде медианы, и граничные значения в виде наименьшего и наибольшего наблюдений:

```python
idx = sorted(center_spb_data['rooms'].unique())
spb_observation_rooms = pd.DataFrame(index=idx, columns=['median_price_mln_r', 'largest_observation_mln_r', 'smallest_observation_mln_r'])
for room, group_data in center_spb_data.groupby('rooms'):
    spb_observation_rooms.loc[room, 'median_price_mln_r'] = round(group_data["last_price"].median(), 2)
    spb_observation_rooms.loc[room, 'largest_observation_mln_r'] = round(get_observation(group_data["last_price"]), 2)    
    spb_observation_rooms.loc[room, 'smallest_observation_mln_r'] = round(get_observation(group_data["last_price"], 'smallest'), 2)
display(spb_observation_rooms)
```
